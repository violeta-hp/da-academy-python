### Exam
Let's create our own encryption method, this will be a simple character substitution cipher that replaces a character with the "7th" character after it in the Unicode Standard. 


Write a function called "encrypt_decrypt" that takes 3 parameters:
1. "an_string": String (this will be encrypted/decrypted)
2. "encryption_factor": integer (this will be used for encrypt/decrypt, use "7" as default)
3. "encryption": boolean (when "True" then "encrypt", else "decrypt")

So, when the third parameter is "True", the function must return a new string that contains the characters from the original string, but "replaced" with the "7th" character after it in the Unicode Standard. 

When the third parameter is "False", the function must return a "decrypted" string  (the "7th" character before it in the Unicode Standard)

Example: 
* Unicode for "a" is 97, and  97 + 7 = 104, which corresponds to "h". 
* Following this rule, when "encryption" is "True", the string "abcd" must be encrypted as "hijk". 
* However, if "encryption" is "False" AND the string is "hijk", then the function must return "abcd" (decrypted string).

Hint: You might want to use the built-in functions ord, which converts a character to a numeric code, and chr, which converts numeric codes to characters.
reference: https://www.journaldev.com/22767/python-ord-chr

#### Question: decrypt the following message: "oh}l'h'nvvk'~llrluk" ####



```python
# Your answer here:

def encrypt_decrypt(an_string= "hola", encryption_factor=7, encryption=True):
    letter=[]
    for i in an_string: 
        if encryption: 
            letter.append(chr(ord(i)+ encryption_factor))
        else: 
            letter.append(chr(ord(i)-encryption_factor))
    return "".join(letter)
encrypt_decrypt("oh}l'h'nvvk'~llrluk", encryption=False)

#assert encrypt_decrypt("abcd") == "hijk", "There's something wrong with your function"
#assert encrypt_decrypt("have",encryption_factor=2) == "jcxg", "There's something wrong with your function"
#assert encrypt_decrypt("~llrluk",encryption=False) == "weekend", "There's something wrong with your function"
#assert encrypt_decrypt("hijk",encryption=False) == "abcd",  "There's something wrong with your function"
```

```python
def encrypt_decrypt(an_string= "hola", encryption_factor=7, encryption=True):
    if encryption: 
        return "".join([chr(ord(i)+ encryption_factor) for i in an_string])
    else: 
        return "".join([chr(ord(i)-encryption_factor) for i in an_string])

encrypt_decrypt("oh}l'h'nvvk'~llrluk", encryption=False)
```
